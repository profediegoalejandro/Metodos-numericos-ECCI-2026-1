"""
Actividad: Encontrando la raíz de f(x) = x^3 - x - 2
Usando el método de bisección en Python (Google Colab)

Pasos a desarrollar:

1. Definir la función f(x) usando def.
2. Inicializar el intervalo:
       x_l = 1, x_u = 2
3. Implementar el ciclo while con criterio de convergencia:
       |x_u - x_l| < 0.01
4. Calcular el punto medio:
       x_r = (x_l + x_u)/2
5. Evaluar condición para actualizar el intervalo:
       f(x_l)*f(x_r) < 0
6. Mostrar por iteración:
       - Número de iteración
       - x_l, x_u, x_r
       - f(x_r)
       - |x_u - x_l|
7. Reportar raíz aproximada y número de iteraciones.
8. Graficar la función y marcar la raíz usando matplotlib.
"""






````python
# ---------------------------------
# MÉTODO DE BISECCIÓN
# f(x) = x^3 - x - 2
# Criterio: |xu - xl| < 0.01
# ---------------------------------

# 1) Definir la función
def f(x):
    return x**3 - x - 2

# 2) Inicializar intervalo
xl = 1
xu = 2

# 3) Tolerancia
tol = 0.01

# 4) Contador
iteracion = 0

print("Iter   xl        xu        xr        f(xr)      |xu-xl|")

# Verificar cambio de signo
if f(xl) * f(xu) > 0:
    print("No existe cambio de signo en el intervalo")
else:

    # 5) Ciclo while
    while abs(xu - xl) >= tol:

        iteracion += 1

        # Punto medio
        xr = (xl + xu) / 2

        # Mostrar información
        print(f"{iteracion:<6} {xl:.4f}   {xu:.4f}   {xr:.4f}   {f(xr):.4f}   {abs(xu-xl):.4f}")

        # Cambio de signo
        if f(xl) * f(xr) < 0:
            xu = xr
        else:
            xl = xr

    # Raíz final
    raiz = (xl + xu) / 2

    print("\nRaíz aproximada:", round(raiz,6))
    print("Número de iteraciones:", iteracion)
````

````python
import numpy as np
import matplotlib.pyplot as plt

x_vals = np.linspace(0,3,400)
y_vals = f(x_vals)

plt.plot(x_vals, y_vals)
plt.axhline(0)
plt.scatter(raiz, f(raiz))
plt.title("Método de Bisección")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid()
plt.show()
````

````python
#-------------------------------------------------
# Actividad: Encontrando la raíz de f(x) = x^3 - x - 2
# Usando el método de bisección en Python (Colab)
#-------------------------------------------------

import matplotlib.pyplot as plt

# 1. Definir la función f(x)
def f(x):
    return x**3 - x - 2

# 2. Inicializar el intervalo
x_l = 1
x_u = 2  # Ojo: normalmente el intervalo para f(x)=x^3-x-2 es [1,2]

# 3. Criterio de convergencia
tolerancia = 0.01

# Variables para el ciclo
iteracion = 0
x_r = (x_l + x_u)/2

# Listas para graficar la evolución de x_r y f(x_r)
xr_list = []
fxr_list = []

# 4. Ciclo while de bisección
while abs(x_u - x_l) >= tolerancia:
    iteracion += 1
    x_r = (x_l + x_u)/2
    fx_r = f(x_r)

    # Guardar valores para graficar
    xr_list.append(x_r)
    fxr_list.append(fx_r)

    # 5. Mostrar información de la iteración
    print(f"Iteración {iteracion}: x_l={x_l:.4f}, x_u={x_u:.4f}, x_r={x_r:.4f}, f(x_r)={fx_r:.4f}, |x_u-x_l|={abs(x_u-x_l):.4f}")

    # 6. Actualizar el intervalo
    if f(x_l) * fx_r < 0:
        x_u = x_r
    else:
        x_l = x_r

# 7. Reportar raíz aproximada y número de iteraciones
print("\nRaíz aproximada:", x_r)
print("Número de iteraciones:", iteracion)

# 8. Graficar la función y marcar la raíz
x_vals = [i/10 for i in range(10, 21)]  # de 1 a 2
y_vals = [f(x) for x in x_vals]

plt.figure(figsize=(8,5))
plt.plot(x_vals, y_vals, label="f(x)")
plt.axhline(0, color='black', linewidth=0.8)
plt.scatter(x_r, f(x_r), color='red', label=f"Raíz ≈ {x_r:.4f}")
plt.title("Método de Bisección - f(x) = x^3 - x - 2")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid(True)
plt.show()
````

````python
# Actividad: Encontrando la raíz de f(x) = x^3 - 4x - 9
# Usando el método de Falsa Posición en Python (Colab)

import matplotlib.pyplot as plt

# 1. Definir la función f(x)
def f(x):
    return x**3 - 4*x - 9

# 2. Inicializar el intervalo
xl = 2
xu = 3

# 3. Inicialización del ciclo
xr = xu  # valor anterior de xr
epsilon = 0.01
error = 1.0
iteracion = 0

# Listas para graficar
xr_list = []
iter_list = []

# 4. Ciclo while hasta que |xr(k) - xr(k-1)| < 0.01
while error > epsilon:
    iteracion += 1
    xr_new = xu - (f(xu)*(xl - xu)) / (f(xl) - f(xu))

    # 5. Calcular error
    error = abs(xr_new - xr)

    # 6. Mostrar iteración
    print(f"Iter {iteracion}: xl={xl:.4f}, xu={xu:.4f}, xr={xr_new:.4f}, f(xr)={f(xr_new):.4f}, error={error:.4f}")

    # 7. Actualizar el intervalo según f(xl)*f(xr)
    if f(xl) * f(xr_new) < 0:
        xu = xr_new
    elif f(xl) * f(xr_new) > 0:
        xl = xr_new
    else:
        error = 0  # Raíz exacta encontrada

    xr = xr_new
    xr_list.append(xr)
    iter_list.append(iteracion)

# 8. Reportar raíz aproximada y número de iteraciones
print(f"\nRaíz aproximada: x ≈ {xr:.4f} después de {iteracion} iteraciones.")

# 9. Graficar función y marcar raíz
x = [i/100 for i in range(0, 500)]
y = [f(val) for val in x]

plt.plot(x, y, label='f(x)')
plt.axhline(0, color='black', linewidth=0.7)
plt.plot(xr, f(xr), 'ro', label=f'Raíz ≈ {xr:.4f}')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Método de Falsa Posición')
plt.legend()
plt.grid(True)
plt.show()
````
